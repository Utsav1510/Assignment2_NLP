```python
# Define predefined categories
categories = {
    "Text Preprocessing": "Tokenization, stopword removal, cleaning",
    "Model Training": "Training errors, overfitting, validation, epochs",
    "Tool Usage": "Using libraries like spaCy, NLTK, HuggingFace",
    "Implementation Issue": "Bug reports, installation, environment setup",
    "Sentiment Analysis": "Posts about sentiment prediction/classification",
    "NER / Text Classification": "Named Entity Recognition or doc classification",
    "Language Models": "Using/understanding models like GPT, BERT",
    "Conceptual Understanding": "Theoretical differences, model architecture doubts",
    "Other": "Anything not clearly fitting above"
}
```

```python
def categorize_post(title):
    title = title.lower().strip()

    # Text Preprocessing
    if any(kw in title for kw in ["tokenize", "punctuation", "lemmatize", "stemming", "remove stopword", "preprocess"]):
        return "Text Preprocessing"

    # Model Training
    elif any(kw in title for kw in ["train", "fit", "overfit", "epoch", "loss", "accuracy", "validation", "split data"]):
        return "Model Training"

    # Tool Usage
    elif any(kw in title for kw in ["spacy", "nltk", "transformers", "huggingface", "gensim"]):
        return "Tool Usage"

    # Implementation Issue
    elif any(kw in title for kw in ["error", "bug", "install", "version", "not found", "import", "runtime"]):
        return "Implementation Issue"

    # Sentiment Analysis
    elif "sentiment" in title:
        return "Sentiment Analysis"

    # NER / Text Classification
    elif any(kw in title for kw in ["ner", "named entity", "text classification", "classify text", "intent classification"]):
        return "NER / Text Classification"

    # Language Models
    elif any(kw in title for kw in ["gpt", "bert", "llm", "language model", "transformer"]):
        return "Language Models"

    # Conceptual Understanding
    elif any(kw in title for kw in ["difference between", "what is", "how does", "concept", "explain"]):
        return "Conceptual Understanding"

    # Other
    else:
        return "Other"
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv('split_part1_christen_clean_3rd_Try.csv')

# Define the categorization function
def categorize_post(title):
    title = str(title).lower().strip()

    if any(kw in title for kw in ["tokenize", "punctuation", "lemmatize", "stemming", "remove stopword", "preprocess"]):
        return "Text Preprocessing"
    elif any(kw in title for kw in ["train", "fit", "overfit", "epoch", "loss", "accuracy", "validation", "split data"]):
        return "Model Training"
    elif any(kw in title for kw in ["spacy", "nltk", "transformers", "huggingface", "gensim"]):
        return "Tool Usage"
    elif any(kw in title for kw in ["error", "bug", "install", "version", "not found", "import", "runtime"]):
        return "Implementation Issue"
    elif any(kw in title for kw in ["ner", "named entity", "text classification", "classify text", "intent classification"]):
        return "NER / Text Classification"
    elif any(kw in title for kw in ["gpt", "bert", "llm", "language model", "transformer"]):
        return "Language Models"
    else:
        return "Other"

# ✅ Apply the function to each row in the title column
df['predicted_category'] = df['title'].apply(categorize_post)

# Optionally save or display the result
df.to_csv('member1_christen_categorized_3nd_Try.csv', index=False)
print("✅ Categorization applied and saved to 'member1_christen_categorized_3nd_try.csv'")
```

```python
df_c = pd.read_csv("member1_christen_categorized_3nd_Try.csv")
df_c.head()
```

```python
import pandas as pd
import ast



# Convert date columns to datetime
df_c['creation_date'] = pd.to_datetime(df_c['creation_date'], errors='coerce')
df_c['last_activity_date'] = pd.to_datetime(df_c['last_activity_date'], errors='coerce')

# Safely convert stringified tag lists to Python lists
df_c['tags'] = df_c['tags'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else [])

# 1. Number of posts with no answers
no_answer_count = (df_c['answer_count'] == 0).sum()
print(f"Number of posts with no answers: {no_answer_count}")

# 2. Average view count
average_view_count = df_c['view_count'].mean()
print(f"Average view count: {average_view_count:.2f}")

# 3. Tags associated with posts that have no answers
tags_with_no_answers = df_c[df_c['answer_count'] == 0].explode('tags')['tags'].value_counts()
print("\nTop tags from posts with no answers:")
print(tags_with_no_answers)

# 4. Median time to first reply (in hours)
df_c['time_to_first_reply'] = (df_c['last_activity_date'] - df_c['creation_date']).dt.total_seconds() / 3600
median_reply_time = df_c['time_to_first_reply'].median()
print(f"\nMedian time to reply (in hours): {median_reply_time:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the categorized dataset
df = pd.read_csv('member1_christen_categorized_3nd_Try.csv')

# Count posts per category
category_counts = df['predicted_category'].value_counts()

# Plotting the distribution as a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    y=category_counts.values, 
    x=category_counts.index, 
    palette="Set3"
)

# Add titles and labels
plt.title("Number of Posts per Category", fontsize=14)
plt.xlabel("Category")
plt.ylabel("post_count")
plt.tight_layout()

# Show the plot
plt.show()
```
